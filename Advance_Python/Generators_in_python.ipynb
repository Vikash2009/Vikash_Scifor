# **What is A Generator?**
Python generators are simple way of creating iterators,

## **The why**

```python
l=[x for x in range(100000)]

import sys
sys.getsizeof(l)
```

see how much memory does an iterator takes

```python
x=range(100000)

sys.getsizeof(x)
```

the memory occupied is way lesser than iterators

##**A Simple Example**

```python
def gen_demo():

  yield('first statement')
  yield('second statement')
  yield('third statement')
```

```python
gen=gen_demo()

for i in gen:
  print(i)
```

## **Example 2**

```python
def my_range(start,end):

  for i in range(start,end):
    yield i
```

```python
for i in my_range(12,20):
  print(i)
```

## **Generator Expression**

```python
# first lets see by list comprehension
L=[i**2 for i in range(10,20)]
L
```

```python
# by using generator
G=(i**2 for i in range(10,20))

for i in G:
  print(i)
```

## **Advantages of Generators**

**1. Memory Efficient**

```python
L=[x for x in range(100000)]
gen=(x for x in range(100000))

print('Size of L in memory', sys.getsizeof(L))
print('Size of gen in memory', sys.getsizeof(gen))
```

**2. Represnting Infinite Streams**

```python
def all_even_num():
  n=0
  while True:
    yield n
    n+=2
```

```python
even_num_gen=all_even_num()

next(even_num_gen)
next(even_num_gen)
```

**Chaining Generator**

```python
def fibonacci_num(nums):
  x,y=0,1
  for _ in range(nums):
    x, y=y, x+y
    yield x

def square(nums):
  for num in nums:
    yield num**2

print(sum(square(fibonacci_num(10))))
```
