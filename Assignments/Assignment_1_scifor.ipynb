Q1) Write a Python function that takes a list and returns a new list with distinct elements from
the first list.


```python
# defining the function
def new_list(input_list):
  return list(set(input_list))

# usage
input_list=[1,1,2,2,2,3,4,55,55,6,6]
print(new_list(input_list))
```

Q4)  Write a program to print the following star pattern using the for loop


```python
n=5
for i in range(n):
  for j in range(i+1):
    print("* ", end="")
  print()

for i in range(n-1,0,-1):
  for j in range(i):
    print("* ", end="")
  print()
```

Q5) Write a program to accept a number from a user and calculate the sum of all numbers
from 1 to a given number


```python
def add_num(n):
  sum_of_nums=0
  for i in range(1,n+1):
    sum_of_nums += i
  return sum_of_nums

n = int(input("enter the number : "))
sum = add_num(n)

print(f"sum of all numbers from 1 to {n} is" , sum)
```
