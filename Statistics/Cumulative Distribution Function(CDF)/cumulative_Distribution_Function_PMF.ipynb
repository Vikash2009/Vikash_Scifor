# Plotting Graph of Probability Mass Function(PMF)

```python
import pandas as pd
import random
```

```python
# PMF for rolling 2 dices

L=[]
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)
  L.append(a+b)
```

```python
len(L)
```

```python
L[:5]
```

```python
# creating Series of probability of outcomes
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
import numpy as np

# applying cumulative sum
np.cumsum(s)
```

```python
np.cumsum(s).plot(kind='bar')
```

**it gives the probability of X <= x**
