# **removing numbers using regex**

```python
def remove_num(text):
  result = re.sub(r'\d+', ' ', text)
  return result
```

```python
input_string = 'you bought 6 candies from shop, and 4 candies are in home'
remove_num(input_string)
```

# **converting numbers to words using inflect**

```python
import inflect
q = inflect.engine()
def convert_num_word(text):
  temp_string = text.split()
  new_str = []

  for word in temp_string:
    if word.isdigit():
      temp = q.number_to_words(word)
      new_str.append(temp)
    else:
      new_str.append(word)

  temp_string = ' '.join(new_str)
  return temp_string
```

```python
input_str = 'you bought 600 candies from shop, and 4 candies are in home'
convert_num_word(input_str)
```

# **remove punctuations**

```python
def rem_punc(text):
    # Create a translation table: map punctuation to None
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
input_str = 'hey, are you exited !!!, after a week , we will be in shimla !!!'
rem_punc(input_str)
```

# **removing stopwords**

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
```

```python
def rem_stopwords(text):
  stop_words = set(stopwords.words('english'))
  word_tokens = word_tokenize(text)
  filtered_text = [word for word in word_tokens if word not in string.punctuation]
  filtered_text = [word for word in filtered_text if word not in stop_words]
  return filtered_text
```

```python
text = 'data is the new oil. A.I is the last invention'
rem_stopwords(text)
```

# **stemming**

```python
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
stem1 = PorterStemmer()
```

```python
# stem words in the list of tokenize
def s_words(text):
  word_tokens = word_tokenize(text)
  stem = [stem1.stem(word) for word in word_tokens]
  return stem
```

```python
text = 'data is the new revolution in the world , in a day one individual would generate terabytes od data'
s_words(text)
```

# **lemmatization**

```python
from nltk.stem import wordnet
from nltk.tokenize import word_tokenize

lemma = wordnet.WordNetLemmatizer()
nltk.download('wordnet')

def lemmatize_word(text):
  '''if there is more than one word it will convert it in verb(v)'''
  word_tokens = word_tokenize(text)

  # provide context i.e ; part of peech(pos)
  lemmas = [lemma.lemmatize(word, pos ='v')for word in word_tokens]
  return lemmas
```

```python
text = 'data is the new revolution in the world , in a day one individual would generate terabytes od data'
lemmatize_word(text)
```

# **Chunking**

```python
# importing libraries
from nltk.tokenize import word_tokenize
from nltk import pos_tag

# here we define chunking function with text and regular
# expressions representing grammar as parameter
def chunking(text, grammar):
  word_tokens = word_tokenize(text)

  # label words with pos
  word_pos = pos_tag(word_tokens)

  # create chunk parser using grammar
  chunkParser = nltk.RegexpParser(grammar)

  # test it on the list of word tokens with tagged pos
  tree = chunkParser.parse(word_pos)

  # iterate over the parse tree and print subtrees
  for subtree in tree.subtrees():
    print(subtree)

# sentence to be chunked
sentence = 'a white crow is swimming inside river'

# Regular expression grammar for Noun Phrase (NP)
grammar = "NP: {<DT>?<JJ>*<NN>}"

# Calling the chunking function with the sentence and grammar
chunking(sentence, grammar)
```

# **Named Entity Recognition**

```python
#Importing tokenization and chunk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')

def ner(text):                    #tokenize the text
  word_tokens = word_tokenize(text)

  #pos tagging of words
  word_pos = pos_tag(word_tokens)

  #tree of word entities
  print(ne_chunk(word_pos))

text = 'Sachin Tendulkar scored the highest 400 runs in a test match which played in between India and England.'
ner(text)
```

# **re.match()**

```python
import re

lists = ['amazing animals', 'awesome apples', 'astronaut adventure']

for i in lists:
    q = re.match("(a\w+)\W(a\w+)", i)

    if q:
        print(q.groups())
```

# **re.search()**

```python
pattern = ["happy", "day"]
text = "Today is a happy day."

for p in pattern:
    print("You're looking for '%s' in '%s'" % (p, text), end='')

    if re.search(p, text):
        print(' Found match!')
    else:
        print(' No match found!')
```
