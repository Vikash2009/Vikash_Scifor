## **Plotting Using A DataFrame**

```python
data=pd.read_csv('/content/tips.csv')
data.head()
```

## **Bar PLot**

```python
data.groupby('sex')['total_bill'].mean().plot(kind='bar', color=['g', 'c'],figsize=(10,5))
plt.xlabel('Sex')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill by Sex');
```

## **Line Plot**

```python
plt.figure(figsize=(8, 6))
plt.plot(data['total_bill'])
plt.xlabel('Index')
plt.ylabel('Total Bill')
plt.title('Total Bill over Index')
plt.grid(True)
plt.show()
```

## **Scatter PLot**

```python
import seaborn as sns

sns.scatterplot(data=data, x='total_bill', y='tip',color='purple')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip')
plt.show()
```

## **Stacked Line Plot**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data[['total_bill', 'tip']], dashes=False)
plt.xlabel('Index')
plt.ylabel('Amount')
plt.title('Total Bill and Tip over Index')
plt.legend(title='Features', loc='upper left', labels=['Total Bill', 'Tip'])
plt.show()
```

## **Pie Plot**

```python
plt.figure(figsize=(8, 6))
data['sex'].value_counts().plot(kind='pie',explode=(0,0.1) ,startangle=90, colors=['blue', 'orange'])
plt.title('Distribution of Sex')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data[['total_bill', 'tip']])
plt.xlabel('Index')
plt.ylabel('Amount')
plt.title('Total Bill and Tip over Index')
plt.legend(title='Features', labels=['Total Bill', 'Tip'])
plt.show()
```

```python
plt.scatter(data=data, x='total_bill', y='tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs. Tip')
plt.show()
```

## **Stacked Bar Chart**

```python
plt.figure(figsize=(10, 6))
plt.bar(data.index, data['total_bill'], label='Total Bill', color='skyblue')
plt.bar(data.index, data['tip'], bottom=data['total_bill'], label='Tip', color='orange')
plt.xlabel('Index')
plt.ylabel('Amount')
plt.title('Stacked Bar Chart of Total Bill and Tip')
plt.legend()
plt.show()
```

```python
sns.lineplot(x='total_bill', y='tip', data=data);
```
