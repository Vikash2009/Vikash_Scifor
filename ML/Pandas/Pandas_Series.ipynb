## **Imporing Pandas**

```python
import pandas as pd
import numpy as np
```

## **Series from list**

```python
# string
country=['india','nepal','BD','srilanka']

pd.Series(country)
```

```python
# integers
runs=[10,50,20,80,100]

pd.Series(runs)
```

```python
# custom index
marks_series=[50,60,70,100]
subjects=['english','maths','bio','DS']

pd.Series(marks_series,index=subjects)
```

```python
# setting a name
marks=pd.Series(marks_series,index=subjects,name='vikash ke marks')
marks
```

## **Series From Dictionary**

```python
marks_dict={'maths':50,
       'bio':70,
       'geo':80,
       'DS':100}

pd.Series(marks_dict)
```
